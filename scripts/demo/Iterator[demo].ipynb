```python
# iter() --> __iter__()
# Rule : return the iterator object itself

# next() --> __next__()
# Rule : return the next item or raise StopIterator error
```

```python
class Student:
    def __init__(self, id, name, marks):
        self.seq = id
        self.id = id 
        self.name = name 
        self.marks = marks 
        
    def __iter__(self):
        #print("*** __iter__() is called***")
        return self
        
    def __next__(self):
        #print("*** __next___() is called***")
        if self.seq < 60:
            self.seq = self.seq + 1
            return self.seq
        else:
            raise StopIteration
```

```python
for n in Student(49, "venkat", 100):
    print(n)
    
```

```python
class MyIterator:
    def __init__(self):
        self.num = 0
        
    def __iter__(self):
        print("*** __iter__() is called ***")
        return self

    def __next__(self):
        print("*** __next__() is called -- start ***")
        if self.num < 10:
            self.num = self.num + 1
            return self.num
        else:
            raise StopIteration
        print("*** __next__() is called -- end ***")
```

```python
for item in MyIterator():
    print(item)
```

```python
names = ["anusha", "bharath", "pavan", "naveen"]
```

```python
for name in names:
    print(name)
```

```python
names_iterator = iter(names)

while(True):
    try:
        name = next(names_iterator)
        print(name)
    except StopIteration:
        break
```

```python
mi = MyIterator()
it = iter(mi)
while(True):
    try:
        item = next(it)
        print(item)
    except StopIteration:
        break
```
