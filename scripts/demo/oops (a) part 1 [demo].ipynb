```python
class Student:
    
    # __init__ for costructing the object and initializing the attributes of the object 
    def __init__(self, x, y):
       self.id = x
       self.name = y
       self.marks = z

    # string representation of the object
    def __str__(self):
        return f"id={self.id}, name={self.name}, marks={self.marks}"

s1 = Student(1, "anusha", 99)
s2 = Student(2, "naveen", 98)
s3 = Student(3, "gowtham", 100)
s4 = Student(4, "pavan", 97)
s5 = Student(5, "jaggu", 80)
s6 = Student(6, "hemu", 100)

students = [s1, s2, s3, s4, s5, s6]

for std in students:
    print(std)
```
