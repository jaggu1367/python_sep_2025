```python
# windows encoding supports 0 - 255 , range of 256 characters only
file_path = "D:/work/bigdata/python_demo_project/my_data.txt"
with open(file_path) as f:
    data = f.read()
    print(data)
```

```python
file_path = "D:/work/bigdata/python_demo_project/my_data.txt"
with open(file_path, encoding="ascii") as f:
    data = f.read()
    print(data)
```

```python
# 0 , 255 ~ 256 characters supported by cp1252( Windows Western Europe Character Encoding)
file_path = "D:/work/bigdata/python_demo_project/my_data.txt"
with open(file_path, encoding="cp1252") as f:
    data = f.readline()
    while(data !=''):
        data = f.readline()
        print(data)
```

```python
# lakhs of characters supported by utf-8 (436K and so on)
file_path = "D:/work/bigdata/python_demo_project/my_data.txt"
with open(file_path, encoding="utf-8") as f:
    data = f.read()
    print(data)
```
