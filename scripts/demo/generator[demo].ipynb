```python
def big_numbers():
    
    i = 1 
    while(i<=10):
        yield i
        i = i + 1

for num in big_numbers():
    print(num)
```

```python
# yield using send

def demo():
    msg = yield "Hello"
    msg2 = yield msg.upper()
    yield msg2.capitalize()
```

```python
g = demo()
```

```python
print(next(g))
```

```python
g.send("hi") # --> send and yild the next value
```

```python
g.send("hello how are you doing")
```
