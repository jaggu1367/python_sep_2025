```python
class Address:
    def __init__(self, door_no, house_no, place):
        self.door_no = door_no 
        self.house_no = house_no
        self.place = place
```

```python
class Student(Person):
   
    def __init__(self, id, name, marks):
        self.id = id
        self.name = name 
        self.marks = marks
        self.address = Address("1", "A", "Ameerpet")

    def __str__(self):
        return f"""
            id = {self.id}
            name = {self.name}
            marks = {self.marks}
        """
```

```python
s1 = Student(1, "anil", 99)
s2 = Student(2, "sunil", 98)
```

```python
# Student IS a Person (Inheritance)
a1 = s1.address # Student HAS a Address (Composition)
```

```python
print(a1.door_no)
print(a1.house_no)
print(a1.place)
```

```python
# Student is a Person
# Teacher is a Person 
# Inheritance : IS a relationship
```

```python
class Car:
    def __init__(self, name):
        print("Car() __init__() method is called")
        self.tyres = 4
        self.doors = 4
        self.name = name
    
    def __str__(self):
        return f"""
            name = {self.name}
        """

    def start(self):
        print("Car() : start method is called")
        
    def run(self):
        print("Car() : run method is called")
        
    def stop(self):
        print("Car() : stop method is called")

    
```

```python
class Tata(Car):
    
    def __init__(self, model):
        super().__init__(name="TATA")
        print("Tata() __init__() method is called")
        self.model = model
        self.safety_rating = 5

    # override parent definition
    def __str__(self):
        return f"""
            name = {self.name},
            model = {self.model}
            doors = {self.doors},
            tyres = {self.tyres},
            safety_rating = {self.safety_rating}
        """
    # override
    def start(self):
        print("Tata() : start method is called")
    
    # override
    def run(self):
        print("Tata() : run method is called")

    # override
    def stop(self):
        print("Tata() : stop method is called")
    
    def sound(self):
        print("Tata() : sound is heard")
```

```python
t1 = Tata("Indica") 
t2 = Tata("Harrier")
t3 = Tata("Safari")
t4 = Tata("Curvv")
```

```python
print(t1) #--> print(str(t1)) --> print(t1.__str__())
print(t2)
print(t3)
print(t4)
```

```python
class Animal:
    def sound(self):
        print("Animals make sounds")

class Dog(Animal):
    def sound(self):
        super().sound()   # Call parent method
        print("Dog adds: Woof Woof")
```

```python
d = Dog()
d.sound()
```

```python
class Bird:
    def sound(self):
        print("Birds make sounds")

class Sparrow(Bird):
    def sound(self):
        print("Sparrow says: Chirp Chirp")

class Crow(Bird):
    def sound(self):
        print("Crow says: Caw Caw")
```

```python
b = Bird()
s = Sparrow()
c = Crow()
```

```python
birds = [b, s, c]
```

```python
for bird in birds:
    bird.sound()
```
