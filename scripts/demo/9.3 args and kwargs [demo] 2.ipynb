```python
num_list = [12, 15, 28, 46, 33, 27, 48, 49, 98, 99]
```

```python
for x in num_list:
    print(x)
```

```python
even_list = [x for x in num_list if x%2==0]
```

```python
odd_list = [x for x in num_list if x%2==1]
```

```python
num_list
```

```python
even_list
```

```python
odd_list
```

```python
def message(*args):
    values = args
    print(values)
    print(len(values))
```

```python
message(num_list, even_list, odd_list)
```

```python
message(*num_list, even_list, odd_list)
```

```python
message(*num_list)
```

```python
message(*num_list, *even_list, *odd_list)
```

```python
print(num_list)
```

```python
print(*num_list)
```

```python
message(even_list)
```

```python
def message(*args, **kwargs):
    print(type(args))
    print(args)
    print(type(kwargs))
    print(kwargs)
```

```python
message(10, 20, 30 , 40, 50, {"id":1, "name":"anusha", "marks":99})
```

```python
message(1, "anusha", 99, id=2, name="gowtham", marks=100)
```

```python
message(1, "anusha", 99, 777, id=2, name="gowtham", marks=100)
```

```python
def message(*args, **kwargs):
    print(args[1])
    print(kwargs["name"])
```

```python
message(1, "anusha", 99, 777, id=2, name="gowtham", marks=100)
```
