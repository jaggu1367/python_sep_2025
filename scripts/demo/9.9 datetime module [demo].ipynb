```python
class Student:
    def __init__(self, x, y, z):
        self.id = x
        self.name = y
        self.marks = z

    def __str__(self):
        return f"id={self.id},name={self.name},marks={self.marks}"

    def __lt__(self, other):
        return self.marks < other.marks 
        
    def __eq__(self, other):
        return self.marks == other.marks 
```

```python
s1 = Student(1, "anusha", 100)
s2 = Student(2, "naveen", 99)
s3 = Student(3, "venkat", 98)
s4 = Student(4, "gowtham", 100)
```

```python
s1 < s2
```

```python
s4 > s2
```

```python
s1 == s4
```
