```python
class Student:

    def __init__(self, x, y, z):
        self.id = x
        self.name = y
        self.marks = z
```

```python
marks_list = [99, 98, 100, 100, 95, 97]
```

```python
s1 = Student(1, "venkat", marks_list)
```

```python
import copy

s2 = copy.copy(s1)
```

```python
print(s1.id)
print(s1.name)
print(s1.marks)
```

```python
print(s2.id)
print(s2.name)
print(s2.marks)
```
