```python
class Student():
    def __init__(self, x, y, z):
        #print("_init__() is called")
        self.id = x
        self.name = y 
        self.marks = z

    def __str__(self):
        return f"Student: id={self.id}, name={self.name}, marks={self.marks}"
```

```python
s1 = Student(10, "jagadeesh", 100)
s2 = Student(11, "sukanya", 90)
s3 = Student(12, "anusha", 99)
```

```python
students = [s1, s2, s3]
```

```python
for s in students:
    print(f"id={s.id},name={s.name},marks={s.marks}")
```

```python
# use list comprehension to create a list of marks
# [100, 90, 99]

# [<part1> <part2> <part3>]
# [<expression> <for declaration> <if condition>]

[s for s in students]
```

```python
[s.marks for s in students]
```

```python
[s.marks for s in students if s.marks<100]
```

```python
# use list comprehension to get sum of marks
marks_list = sum([s.marks for s in students])
print(marks_list)
```

```python
#reduce(function, iterable)
from functools import reduce

def add(x,y):
    return x+y

reduce(add, marks_list)
```

```python
from functools import reduce

reduce(lambda x,y:x+y, marks_list)
```

```python
reduce(sum, marks_list)
```

```python
class Voter:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print("Voter Name: ", self.name)
        print("Voter age: ", self.age)

    def update_age(self, new_age):
        self.age = new_age

    def __str__(self):
        #print("__str__ is getting called")
        return f"Voter : name={self.name}, age={self.age}"
        
    def __repr__(self):
        #print("__repr__ is getting called")
        return f"Voter({self.name},{self.age})"
```

```python
v = Voter("sunny", 23)
```

```python
repr(v)
```

```python
print(v)
```

```python
str(v)
```
