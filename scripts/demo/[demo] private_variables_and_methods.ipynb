```python
class Student: # name of the class

    school_name = "ZPHS Korutla" # class variable 
    
    def __init__(self, x, y, z, no): # __init__ instance method, [x, y, z method arguments]
        print("__init__() is called")
        self.id = x              # id instance variable but it is public
        self.name = y            # name instance variable but it is public
        self.marks = z           # marks instance variable but it is public
        self.__phone = no        # __phone instance variable but it is private

    def __str__(self):  # __str__ instance method
        print("__str__() is called")
        return f"id={self.id}, name={self.name}, marks={self.marks}"
        
    def __repr__(self): # __repr__ instance method
        print("__repr__() is called")
        return f"Student({self.id}, {self.name}, {self.marks})"

    # public method
    def display(self): # display instance method
        print("display() is called")
        
        print(f"""Student[id={self.id}, 
        name={self.name},
        marks={self.marks}, 
        phone={self.__phone}, 
        school_name={Student.school_name}]
        """)
        #self.__show()

    # getter method should be public
    def get_phone(self):
        return self.__phone
        
    # setter method should be public
    def set_phone(self, new_phone):
        self.__phone = new_phone
        
    def update_marks(self, x): # update_marks instance method, x -> method argument
        self.marks = x         # marks instance variable 

    # private method
    def __show(self):
        print("show() is called")
        print(f"""Student[id={self.id}, 
        name={self.name},
        marks={self.marks}, 
        phone={self.__phone}, 
        school_name={Student.school_name}, 
        school_type=MAGICAL_SCHOOL
        ]""")

    # getters and setters

    # class method ? 
    # 1) created using a decorator @classmethod
    # 2) instance method uses "self", class method uses "cls"
    # 3) class method is used to access and modify the class variables

    @classmethod
    def update_school_name(cls, x): # update_school_name: class method name, cls: class itself, new_school: method argument
        cls.school_name = x

    @staticmethod
    def add(x, y): # add is a static method, decorated with @staticmethod and does not access class or instance variables
        return x+y
```
