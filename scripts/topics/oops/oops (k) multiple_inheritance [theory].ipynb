# **Multiple Inheritance (Simple & Safe)**

A class can inherit **features from more than one parent class**.

```
class Child(Parent1, Parent2):
    pass
```

## 1. Basic Multiple Inheritance Example

````python
class Father:
    def skill_father(self):
        print("Father: Knows Driving")

class Mother:
    def skill_mother(self):
        print("Mother: Knows Cooking")

class Child(Father, Mother):
    def skill_child(self):
        print("Child: Knows Programming")

c = Child()
c.skill_father()   # From Father
c.skill_mother()   # From Mother
c.skill_child()    # Child's own skill
````
