### Question 1
Use `map()` with `lambda` to divide each value in the list by 3.

**Input:**  
nums = [12, 17, 5, 30, 21]

**Expected Output:**  
[4.0, 5.666666666666667, 1.6666666666666667, 10.0, 7.0]

```python
# Pavan
nums = [12, 17, 5, 30, 21]

list(map(lambda x:x/3, nums))
```

### Question 3
Use `map()` with `lambda` on dictionary items to generate strings of format "Name:Value".

**Input:**  
sales = {"Alice":1500, "Bob":2300, "Charlie":1800}

**Expected Output:**  
["Alice#1500", "Bob#2300", "Charlie#1800"]

```python
# Naveen
sales = {"Alice":1500, "Bob":2300, "Charlie":1800}

list(map(lambda x: f"{x[0]}#{x[1]}", sales.items()))
```

### Question 5
Use `map()` with `lambda` to sum two lists element-wise.

**Input:**  
a = [1,2,3]  
b = [4,5,6]

**Expected Output:**  
[5, 7, 9]

```python
# Anusha
a = [1,2,3]
b = [4,5,6]

# for x in zip(a,b):
#     print(x)
#     print(type(x))
# for x,y in zip(a,b):
#     print(x+y)

list(map(lambda x:x[0]+x[1], zip(a,b)))
```
